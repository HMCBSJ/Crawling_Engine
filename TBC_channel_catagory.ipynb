```
base_url = "https://www.tbc.net.tw/Epg/AllChannel/"

# chrome창(웹드라이버) 열기 
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
# TBC 접속하기 
driver.implicitly_wait(3)
driver.get(base_url)
```

```
#https://stackoverflow.com/questions/25050311/extract-first-item-of-each-sublist
#https://pandas.pydata.org/pandas-docs/version/0.20/merging.html
#https://blog.udemy.com/how-to-create-pandas-dataframes-a-hands-on-guide/
#https://blog.finxter.com/python-list-of-lists/ 맨밑
cha_cate = []
for cha in driver.find_elements_by_class_name('epg'):
  cha_title = []
  title = cha.find_elements_by_xpath('.//td[@class = "chName"]')
  for tit in title:
    cha_title.append(tit.text.split("\n",1))
  cha_cate.append({'category': cha.text.split("\n",1)[0], 'chan_num': list(zip(*cha_title))[0], 
                   'chan_name': list(zip(*cha_title))[1]})
# creat Dataframe
concat_cha = []
for i in cha_cate:
  df = pd.DataFrame(i)
  concat_cha.append(df)

channel_category = pd.concat(concat_cha)
#Channel_Category = channel_category.set_index(['category']).sort_index()
#channel_category.groupby('category').agg(['unique'])
```
